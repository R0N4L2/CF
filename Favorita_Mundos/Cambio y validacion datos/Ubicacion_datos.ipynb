```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def graph_naves(df):
    clas = np.sort(df.NAVE.unique())
    plt.figure(figsize=(20,12))
    for i in clas:
        df_nave = df[df.NAVE==i]
        df_nave_x = df_nave.COORDENADAXGLOBAL
        df_nave_y = df_nave.COORDENADAYGLOBAL
        r = np.random.rand()
        b = np.random.rand()
        g = np.random.rand()
        color = (r, g, b)
        plt.scatter(x=df_nave_x,y=df_nave_y,c=[color], label=i)
        plt.legend(loc='upper right')
    plt.show()
```

```python
df = pd.read_pickle('Verificados.pickle')
df_art = pd.read_csv('articuloClasificacionAbastos_24-03-2020.csv')
df_mun = pd.read_excel('mundos_clasificaciones_marzo_2020V1.xlsx')
```

```python
df_art['CODIGOARTICULO'] = df_art.CODIGOARTICULO.apply(str)
df1 = df.merge(df_art,left_on='CODIGOARTICULO',right_on='CODIGOARTICULO',how='left')
df_mun = df_mun.rename(columns={'COD_CLASIFICACION':'CODIGOCLASIFICACION','CLASE ROTACION':'CLASE_ROTACION'})
df2 = df1.merge(df_mun[['CODIGOCLASIFICACION','MUNDO','CLASE_ROTACION']],left_on='CODIGOCLASIFICACION',right_on='CODIGOCLASIFICACION',how='left')
df3 = df2[~df2.CLASE_ROTACION.isnull()].drop_duplicates().reset_index(drop=True)
df3.to_pickle('Nuevo_verificados.pickle')
df = df3.copy()
```
